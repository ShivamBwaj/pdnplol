```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv' , encoding='unicode_escape')
```

```python
df_laptops.head(3)
```

# drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])[['Company','Price_euros']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops=df_laptops.sort_values(['Company','Price_euros'])
```

```python
# cheapest: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')
```

```python
# most expensive: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')
```

```python
# Argument

# inplace: drop duplicates in place or to return a copy.
# ignore_index=True: the resulting axis will be labeled 0, 1, …
df_laptops.drop_duplicates(['Company','Price_euros'],ignore_index=True)
```

Task: Find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
# sort dataframe asending by company and inches
# smallest first and biggest last
df_laptop = pd.read_csv('laptop_price.csv' , encoding='unicode_escape')
df_laptop=df_laptop.sort_values(['Company','Inches'])
```

```python
# smallest: keep='first'
df_laptop.drop_duplicates(['Inches'],keep='first')
```

```python
# biggest: keep='last'
df_laptop.drop_duplicates(['Inches'],keep='last')
```
