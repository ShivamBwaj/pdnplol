```python
import pandas as pd
```

```python
# reading the csv file
df_population_raw = pd.read_csv('population_total.csv',encoding='unicode_escape')
```

```python
# first look at the dataset
df_population_raw
```

# Making a Pivot Table

.pivot(): Returns reshaped DataFrame organized by given index / column values ("pivot without aggregation")

```python
# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
# making a pivot table
df_pivot=df_population_raw.pivot(index='year',
                 columns='country',
                 values='population')
```

```python
# selecting some countries
df_pivot=df_pivot[['India','China']]
```

```python
# showing pivot table
df_pivot
```
