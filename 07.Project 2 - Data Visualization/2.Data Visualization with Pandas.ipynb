```python
import pandas as pd
```

```python
# reading the csv file
df_population_raw = pd.read_csv('population_total.csv',encoding='unicode_escape')
```

```python
# first look at the dataset
df_population_raw
```

# Making a Pivot Table

.pivot(): Returns reshaped DataFrame organized by given index / column values ("pivot without aggregation")

```python
# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
# making a pivot table
df_pivot = df_population_raw.pivot(index='year', columns='country',
                                        values='population')
df_pivot1 = df_population_raw.pivot(index='year', columns='country',
                                        values='population')
```

```python
# selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

# Lineplot

```python
df_pivot.plot(kind='line',xlabel='YEAR',ylabel='Population',title='Population V/S Years',figsize=(8,4))
```

# Barplot

## Single Barplot

```python
# selecting only one year (2020)
df_pivot2020=df_pivot[df_pivot.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot2020=df_pivot2020.T  ##shoould have format for barplot
```

```python
# making barplot
df_pivot2020.plot(kind='bar',color='orange',xlabel='Country',ylabel='Population',title='Title holder')
```

## Barplot grouped by “n” variables

```python
# selecting a few years
df_pivot=df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
# making grouped barplot
df_pivot.plot(kind='bar', xlabel='year', ylabel='Population')
```

# Piechart

```python
# changing column name 
df_pivot2020.rename(columns={2020:'2020'},inplace=True)
```

```python
# making the plot
df_pivot2020.plot(kind='pie',y='2020')
```
